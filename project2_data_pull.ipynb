```python
# Remove Multi-level index
ratio_df_new = df_ratios.reset_index(level="Ticker")
ratio_df_new = ratio_df_new.reset_index(level="Report Date")
ratio_df_new["unique key"] = ratio_df_new["Ticker"]+ratio_df_new["Publish Date"].astype(str)
ratio_df_new
```

```python
# Set "Publish Date" column as the new index
ratio_df_new.set_index('Publish Date', inplace=True)
ratio_df_new
```

```python
# Drop unneceassry columns
    # SimFinID has no relation to stock performance
    # All stock ratios are in US dollars
    # Restated date
    # Dividends Per Share and Dividends Payout Ratio filled with NaN or 0.0
ratio_df_new.drop(columns=['SimFinId', 'Currency', 'Restated Date', 'Dividends Per Share', 'Dividend Payout Ratio'], axis=1, inplace=True)
ratio_df_new
```

```python
# Create list of stocks to create portfolios
    # S&P500 top 2 stocks per sector
    # Dow Jones 30
    # Full S&P 500 
# Load the S&P 500 Sector data
df_sp500_sector = pd.read_csv('top_stocks_per_sector_sp500.csv')
df_sp500_sector.head()
```

```python
# Load the full list of S&P 500 stocks
df_sp500_full = pd.read_csv('SP500_Constituents_List_Jan_2022.csv')
df_sp500_full.head()
```

```python
# Create a dataframe of the portfolio stocks from the financial ratios dataframe
# S&P Sector
df_sp500_sector_financials = pd.merge(ratio_df_new, df_sp500_sector)
df_sp500_sector_financials
```

```python
# Drop columns "Fiscal Year", and "Fiscal Period" from portfolio financial ratios from all merged dataframes
df_sp500_sector_full.drop(['Fiscal Year', 'Fiscal Period'], axis=1, inplace=True)
df_sp500_sector_full
```

```python
# Format the date time
df_sp500_sector_full.index = pd.to_datetime(df_sp500_sector_full.Date)
```

```python
# Groupby Ticker for the S&P500 Sector dataframe so percent change in share price is for each individual stock
df_sp500_sector_full = df_sp500_sector_full.groupby("Ticker").resample("Q").last()
df_sp500_sector_full
```

```python
#Calculate percentage in stock price
df_sp500_sector_full['returns'] = df_sp500_sector_full["Adj. Close"].groupby('Ticker').pct_change().shift(-1)
df_sp500_sector_full
```
